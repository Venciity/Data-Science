# Florentine Families
One famous graph in social network analysis is the "Florentine families" graph. It is implemented by default in networkx.

Each node represents a family, each edge - a marital relationship (i.e. someone from family A married someone from family B). Since marriage is mutual (at least, it should be :D), the graph is undirected.

You can load it and visualize it like this:

```python
g = nx.florentine_families_graph()
nx.draw(g, with_labels = True)
```

We will use this graph in the following questions.

````python
families_graph = nx.florentine_families_graph()
nx.draw_networkx(families_graph, with_labels = True)
plt.show()
````

How many nodes does the shortest path from Pazzi to Peruzzi contain, including the starting and the final one?

````python
nodes = nx.shortest_path(families_graph, "Pazzi", "Peruzzi")
print(nodes)
print(len(nodes))

# Second solution
number_of_edges = nx.shortest_path_length(families_graph, "Pazzi", "Peruzzi")
print(number_of_edges + 1)
````

## A famous family
What is the name of the most important / famous family in this graph?

The most important node maybe is this one which is in combination with the most number of short paths.
We can use betweenness centrality to find this one.

More information about betweenness centrality can be read from <a href="https://en.wikipedia.org/wiki/Betweenness_centrality" >Wikipedia</a>.

````python
max_value = 0.0
most_famous = ""
betweenness_centrality = nx.betweenness_centrality(families_graph)
for key, value in betweenness_centrality.items():
    if float(value) > max_value:
        max_value = value
        most_famous = key

print(most_famous, max_value)
````
